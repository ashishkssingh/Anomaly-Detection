# Import Libraries

```python
# Import general purpose library
import numpy as np
import pandas as pd
import warnings

# Import plotting libraries
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Import implementation libraries
import sesd
from sklearn.covariance import EllipticEnvelope

# Display plots inline
%matplotlib inline
warnings.filterwarnings('ignore')
```

# Import Dataset

```python
data = pd.read_csv("Dataset/response_times.csv")
```

# Informative analysis of data

```python
data.head()
```

```python
data.describe()
```

```python
df = data['av']
```

# 1. Anomaly Detection using Standard Deviation

```python
# Creating a sample dataset for better visualization
test_dataset = df[:300]

# calculate summary statistics
data_mean, data_std = np.mean(df), np.std(df)

# Set a cut off value - currently 2 times the standard deviation
cut_off = data_std * 2

# Calculate the upper and lower limit of cut off
lower, upper = data_mean - cut_off, data_mean + cut_off

# identify outliers
outliers = [x for x in df if x < lower or x > upper]
print('Identified outliers: %d' % len(outliers))

# remove outliers
outliers_removed = [x for x in df if x >= lower and x <= upper]
print('Non-outlier observations: %d' % len(outliers_removed))

marks = []
for i in test_dataset:
    if i in outliers:
        marks.append(i)
    else:
        marks.append(np.NAN)
        
plt.figure(figsize = (20,8))
plt.plot(test_dataset, linestyle = '-', label ='ResponseTime')
plt.plot(marks, 'ro', markersize = "3", label ='Anomaly')

red_patch = mpatches.Patch(color='red', label='Anomaly')
blue_lines = mpatches.Patch(color='blue', label='ResponseTime')

plt.legend(handles=[blue_lines, red_patch])
plt.xlabel('TimeStamps')
plt.ylabel('response times')
plt.show()
```

# 2. Anomaly Detection using Interquartile Range Method

```python
# Creating a sample dataset for better visualization
test_dataset = df[:200]

# calculate interquartile range
q25, q75 = np.percentile(df, 25), np.percentile(df, 75)
iqr = q75 - q25
print('Percentiles: 25th=%.3f, 75th=%.3f, IQR=%.3f' % (q25, q75, iqr))

# calculate the outlier cutoff
cut_off = iqr * 1.5
lower, upper = q25 - cut_off, q75 + cut_off

# identify outliers
outliers = [x for x in df if x < lower or x > upper]
print('Identified outliers: %d' % len(outliers))

# remove outliers
outliers_removed = [x for x in df if x >= lower and x <= upper]
print('Non-outlier observations: %d' % len(outliers_removed))

marks = []
for i in test_dataset:
    if i in outliers:
        marks.append(i)
    else:
        marks.append(np.NAN)

# Set Graph size
plt.figure(figsize = (20,8))

# Plot the time series and the outliers
plt.plot(test_dataset, linestyle = '-')
plt.plot(marks, 'ro', markersize = "3")

# Make Legend markers
red_patch = mpatches.Patch(color='red', label='Anomaly')
blue_lines = mpatches.Patch(color='blue', label='ResponseTime')

# Place the legend on the graph
plt.legend(handles=[blue_lines, red_patch])

# Plot the Labels for the axis
plt.ylabel('response times')
plt.xlabel('Data points')

# Show the plotted graph
plt.show()
```

# 3. Anomaly Detection using Gaussian Distribution

```python
test_data = df[:300]


envelope =  EllipticEnvelope(contamination = 0.15) 
X_train = df.values.reshape(-1,1)
envelope.fit(X_train)
df = pd.DataFrame(df)
df['deviation'] = envelope.decision_function(X_train)
df['anomaly'] = envelope.predict(X_train)
df[df['anomaly'] == -1].head()

outliers = df[df['anomaly'] == -1].index.values
print('Identified outliers: %d' % len(outliers))

outliers_removed = len(df['av']) - len(outliers)
print('Non-outlier observations: %d' % outliers_removed)


marks = []
for i in range(0, len(test_data)):
    if i in outliers:
        marks.append(df['av'][i])
    else:
        marks.append(np.NAN)
        
plt.figure(figsize = (20,8))
plt.plot(test_data, linestyle = '-', label ='ResponseTime')
plt.plot(marks, 'ro', markersize = "3", label ='Anomaly')

red_patch = mpatches.Patch(color='red', label='Anomaly')
blue_lines = mpatches.Patch(color='#62A3C9', label='ResponseTime')

plt.legend(handles=[blue_lines, red_patch])
plt.xlabel('TimeStamps')
plt.ylabel('response times (ms)')
plt.show()
```

# 4. Anomaly Detection using SH-ESD Algorithm

```python
outliers_indices = sesd.seasonal_esd(df,seasonality = 25, hybrid=True, max_anomalies=2500, alpha = 3)
print('Non-outlier observations: %d' % len(outliers_indices))

outliers = []
sorted_outliers_indices = np.sort(outliers_indices)
test_outliers_indices = sorted_outliers_indices[:300]
for idx in test_outliers_indices:
    outliers.append(df[idx])
    
marks = []
for i in df[:300]:
    if i in outliers:
        marks.append(i)
    else:
        marks.append(np.NAN)
        
plt.figure(figsize = (20,8))
plt.plot(df[:300], linestyle = '-', label ='ResponseTime')
plt.plot(marks, 'ro', markersize = "3", label ='Anomaly')

red_patch = mpatches.Patch(color='red', label='Anomaly')
blue_lines = mpatches.Patch(color='#62A3C9', label='ResponseTime')

plt.legend(handles=[blue_lines, red_patch])
plt.ylabel('response times')
plt.xlabel('data points')
plt.show()
```
